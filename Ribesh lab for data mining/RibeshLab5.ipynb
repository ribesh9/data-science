```python
le = LabelEncoder()
for col in df.columns:
    df[col] = le.fit_transform(df[col])
```

```python
frequent_itemsets = apriori(df, min_support=0.2, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5) # min_threshold for confidence

# Print the rules (or analyze them further)
print(rules)
```

```python
# Example: Filtering for rules with high lift
high_lift_rules = rules[rules['lift'] >= 1.2]  # Example threshold
print("\nRules with high lift:")
print(high_lift_rules)

# Example: Filtering for rules involving 'Apple'
apple_rules = rules[(rules['antecedents'].apply(lambda x: 'Apple' in x)) | 
                   (rules['consequents'].apply(lambda x: 'Apple' in x))]
print("\nRules involving Apple:")
print(apple_rules)

# Example: Sorting rules by confidence
sorted_rules = rules.sort_values(by='confidence', ascending=False)
print("\nRules sorted by confidence:")
print(sorted_rules)
```
