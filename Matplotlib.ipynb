```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.show() #We need this in .py file
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show() #We need this in .py file
```

```python
categories = ['A','B','C','D']
values = [10, 20, 15, 5]
plt.bar(categories,values)
#plt.barh(categories,values) #gives us horizontal bars.
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Bar Plot')
plt.show()
```

```python
import numpy as np
```

```python
data = np.random.randn(1000) #Generate random data
plt.hist(data,bins = 30)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
data = np.random.rand(1000) #Generate random data
plt.figure(figsize = (20,20), dpi = 300)
plt.hist(data,bins = 30)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,6,8,10]
a = [2,12,33,24,45]
b = [72,64,36,48,120]
plt.plot(x,y,label = 'Series 1',color = 'red',linestyle = '--', marker = '*')
plt.plot(a,b,label = 'Series 2', marker = 'o')
plt.legend(loc = 'upper right')
plt.title('Line Plot')
plt.show() #We need this in .py file
```

```python
#Axis limit and tricks
plt.plot(x,y)
plt.xlim(0,10)
plt.ylim(0,20)
plt.xticks([0,5,10])
plt.yticks([0,5,20])
plt.show()
```
